```python
# 1. Add a key to a dictionary.
d = {0: 10, 1: 20}
d[2] = 30
print(d)
```

```python
# 2. Concatenate dictionaries to create a new one.
dic1 = {1: 10, 2: 20}
dic2 = {3: 30, 4: 40}
dic3 = {5: 50, 6: 60}
result = {**dic1, **dic2, **dic3}
print(result)
```

```python
# 3. Check whether a given key already exists in a dictionary.
d = {1: 10, 2: 20}
key = 2
print(key in d)
```

```python
# 4. Iterate over dictionaries using for loops.
d = {1: 10, 2: 20, 3: 30}
for key, value in d.items():
    print(f"{key}: {value}")
```

```python
# 5. Generate a dictionary that contains a number (1 to n) in the form (x, x*x).
n = 5
d = {x: x*x for x in range(1, n+1)}
print(d)
```

```python
# 6. Print a dictionary where keys are numbers between 1 and 15 and values are squares of the keys.
d = {x: x**2 for x in range(1, 16)}
print(d)
```

```python
# 7. Merge two Python dictionaries.
d1 = {'a': 100, 'b': 200}
d2 = {'c': 300, 'd': 400}
merged = {**d1, **d2}
print(merged)
```

```python
# 8. Iterate over dictionaries using for loops.
d = {'a': 1, 'b': 2, 'c': 3}
for k, v in d.items():
    print(k, v)
```

```python
# 9. Sum all the items in a dictionary.
d = {'a': 10, 'b': 20, 'c': 30}
print(sum(d.values()))
```

```python
# 10. Multiply all the items in a dictionary.
d = {'a': 2, 'b': 3, 'c': 4}
product = 1
for v in d.values():
    product *= v
print(product)
```

```python
# 11. Remove a key from a dictionary.
d = {'a': 100, 'b': 200}
del d['a']
print(d)
```

```python
# 12. Map two lists into a dictionary.
keys = ['a', 'b', 'c']
values = [1, 2, 3]
d = dict(zip(keys, values))
print(d)
```

```python
# 13. Sort a given dictionary by key.
d = {'c': 3, 'a': 1, 'b': 2}
sorted_d = dict(sorted(d.items()))
print(sorted_d)
```

```python
# 14. Get the maximum and minimum values of a dictionary.
d = {'a': 10, 'b': 20, 'c': 5}
print(max(d.values()), min(d.values()))
```

```python
# 15. Get a dictionary from an object's fields.
class Obj:
    a = 1
    b = 2
d = vars(Obj)
print(d)
```

```python
# 16. Remove duplicates from the dictionary (preserve unique key-value pairs).
d = {'a': 1, 'b': 2, 'c': 1}
unique_d = {k: v for i, (k, v) in enumerate(d.items()) if v not in list(d.values())[:i]}
print(unique_d)
```

```python
# 17. Check if a dictionary is empty or not.
d = {}
print(bool(d))
```

```python
# 18. Combine two dictionaries by adding values for common keys.
d1 = {'a': 100, 'b': 200, 'c': 300}
d2 = {'a': 300, 'b': 200, 'd': 400}
result = {k: d1.get(k, 0) + d2.get(k, 0) for k in set(d1) | set(d2)}
print(result)
```

```python
# 19. Print all distinct values in a dictionary.
data = [{'V': 'S001'}, {'V': 'S002'}, {'VI': 'S005'}, {'VII': 'S005'}, {'V': 'S009'}, {'VIII': 'S007'}]
values = set(d['V'] if 'V' in d else d.get('VI', '') for d in data)
print("Unique Values:", values)
```

```python
# 20. Create and display all combinations of letters, selecting each letter from a different key in a dictionary.
data = {'1': ['a', 'b'], '2': ['c', 'd']}
combinations = [x + y for x in data['1'] for y in data['2']]
print(*combinations, sep='\n')
```

```python
# 21. Find highest 3 values in a dictionary
my_dict = {'item1': 400, 'item2': 300, 'item3': 500, 'item4': 200, 'item5': 600}
result = {k: v for k, v in sorted(my_dict.items(), key=lambda x: x[1], reverse=True)[:3]}
print(result)
```

```python
# 22. Combine values in a list of dictionaries
from collections import Counter
data = [{'item': 'item1', 'amount': 400}, {'item': 'item2', 'amount': 300}, {'item': 'item1', 'amount': 750}]
result = Counter()
for d in data:
    result[d['item']] += d['amount']
print(dict(result))
```

```python
# 23. Create a dictionary from a string (letter count)
s = 'SBJAIN'
letter_count = {char: s.count(char) for char in set(s)}
print(letter_count)
```

```python
# 24. Print a dictionary in table format
from tabulate import tabulate
data = {'item1': 100, 'item2': 200, 'item3': 300}
print(tabulate(list(data.items()), headers=['Key', 'Value'], tablefmt='grid'))
```

```python
# 25. Count values associated with a key
my_dict = {'a': [1, 2, 3, 4, 5, 6], 'b': [7, 8]}
key = 'a'
print(len(my_dict[key]))
```

```python
# 26. Convert a list into a nested dictionary of keys
lst = ['a', 'b', 'c']
nested_dict = {}
current = nested_dict
for item in lst:
    current[item] = {}
    current = current[item]
print(nested_dict)
```

```python
# 27. Sort a dictionary alphabetically
my_dict = {'b': 2, 'a': 1, 'c': 3}
sorted_dict = dict(sorted(my_dict.items()))
print(sorted_dict)
```

```python
# 28. Remove spaces from dictionary keys
my_dict = {'key 1': 10, 'key 2': 20}
new_dict = {k.replace(' ', ''): v for k, v in my_dict.items()}
print(new_dict)
```

```python
# 29. Get the top three items in a shop
shop = {'item1': 45.50, 'item2': 35, 'item3': 41.30, 'item4': 55, 'item5': 24}
top_items = sorted(shop.items(), key=lambda x: x[1], reverse=True)[:3]
for item, price in top_items:
    print(item, price)
```
