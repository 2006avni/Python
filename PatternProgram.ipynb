```python
# Task-1 Patterns (examples from the image)
# Pattern 1 (first box: 1, 12, 123, 1234, 12345)
n = 5
num = ""
for i in range(1, n+1):
    num += str(i)
    print(num)
```

```python
# Pattern 2 (second box: 1, 22, 333, 4444, 55555)
n = 5
for i in range(1, n+1):
    print(str(i) * i)
```

```python
# Pattern 3 (third box: 1, 21, 321, 4321, 54321)
n = 5
for i in range(1, n+1):
    s = ""
    for j in range(i, 0, -1):
        s += str(j)
    print(s)
```

```python
# Pattern 4 (fourth box: 1, 101, 10101)
n = 3
for i in range(1, n+1):
    s = ""
    for j in range(i):
        if j % 2 == 0:
            s += "1"
        else:
            s += "0"
    print(s)
```

```python
# Pattern 5 (fifth box: 2, 46, 81012, 14161820)
n = 4
start = 2
for i in range(n):
    s = ""
    for j in range(i+1):
        s += str(start)
        start += 2
    print(s)
```

```python
# Pattern 6 (sixth box: , *, *, *, **)
n = 5
for i in range(1, n+1):
    print("*" * i)
```

```python
# Task-2 Patterns (examples from the image)
# Pattern 1 (first box: A, AB, ABC, ABCD, ABCDE)
n = 5
for i in range(n):
    s = ""
    for j in range(i+1):
        s += chr(65 + j)
    print(s)
```

```python
# Pattern 2 (second box: #, ##, ###, ####, #####)
n = 5
for i in range(1, n+1):
    print("#" * i)
```

```python
# Pattern 3 (third box: p, py, pyt, pyth, pytho, python)
word = "python"
n = len(word)
for i in range(n):
    s = ""
    for j in range(i+1):
        s += word[j]
    print(s)
```

```python
# Task-3
Print numbers from 1 to n as a string without spaces.
n = int(input())
s = ""
for i in range(1, n+1):
    s += str(i)
print(s)
```

```python
# Task-4
Print the pattern:
***
**
*
**
*
n = 5
for i in range(n, 0, -1):
    print("*" * i)
```

```python
# Task-5
Print the pattern (a triangle of asterisks):
    *
   *
  ***
 ***
***
n = 5
for i in range(n):
    print(" " * (n-i-1) + "*" * (2*i+1))
```

```python
# Task-6
Print the star pattern (right-angled triangle decreasing):
***
**
*
**
*

n = 5
for i in range(n, 0, -1):
    print("*" * i)
```

```python
# Task-7
Prime Number Finder between two numbers.
start = int(input("Enter start number: "))
end = int(input("Enter end number: "))
while start <= end:
    is_prime = True
    for i in range(2, start):
        if start % i == 0:
            is_prime = False
            break
    if is_prime and start > 1:
        print(start)
    start += 1
```
