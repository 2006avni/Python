```python
# 1. Electricity bill calculation
# Input customer details
customer_number = input("Enter customer number: ")
units = int(input("Enter units consumed: "))

# Calculate bill amount
if units <= 100:
    amount = units * 1.25
elif units <= 300:
    amount = 100 + (units - 100) * 1.50
elif units <= 500:
    amount = 350 + (units - 300) * 1.75
else:
    amount = 650 + (units - 500) * 2.00
# Output bill
print("\nElectricity Bill")
print("Customer Number:", customer_number)
print("Units Consumed:", units)
print("Amount to Pay: Rs.", amount)
```

```python
# 2. Find the largest number among three numbers
# Input three numbers
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

# Find largest number
largest = num1
if num2 > largest:
    largest = num2
if num3 > largest:
    largest = num3
# Output result
print("Largest number is:", largest)
```

```python
# 3. Python programs
#### a) Add two numbers
# Input two numbers
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Add numbers
sum = num1 + num2
# Output result
print("Sum of", num1, "and", num2, "is:", sum)
```

```python
# b) Find area of a circle
# Input radius
radius = float(input("Enter radius of the circle: "))
pi = 3.14159

# Calculate area
area = pi * radius * radius

# Output result
print("Area of the circle is:", area)
```
