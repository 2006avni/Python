```python
#Lab Assignment-1 Solution (Python)
def convert_to_uppercase(input_file, output_file):
    with open(input_file, 'r') as file:
        content = file.read().upper()
    with open(output_file, 'w') as file:
        file.write(content)
    print(f"Contents written to {output_file} in uppercase.")

# Usage
input_filename = input("Enter input file name: ")
output_filename = input("Enter output file name: ")
convert_to_uppercase(input_filename, output_filename)
```

```python
#Lab Assignment-2 Solution (Python)
def copy_python_script(source_file, dest_file):
    with open(source_file, 'r') as src:
        content = src.read()
    with open(dest_file, 'w') as dest:
        dest.write(content)

    print("Source file content:")
    print(content)
    print(f"\nContent copied to {dest_file}.")
    print("Destination file content:")
    with open(dest_file, 'r') as dest_read:
        print(dest_read.read())
```
