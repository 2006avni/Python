```python
### Lab Assignment-1: Employee & Manager Classes
class Employee:
    def _init_(self, name, age, salary, address):
        self.name = name
        self.age = age
        self.salary = salary
        self.address = address

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Salary: {self.salary}")
        print(f"Address: {self.address}")

class Manager(Employee):
    def _init_(self, name, age, salary, address):
        super()._init_(name, age, salary, address)

    def get_input(self):
        self.name = input("Enter name: ")
        self.age = int(input("Enter age: "))
        self.salary = float(input("Enter salary: "))
        self.address = input("Enter address: ")
```

```python
# Process information of 10 managers
managers = []
for _ in range(10):
    m = Manager("", 0, 0.0, "")
    m.get_input()
    managers.append(m)

print("\nManager Details:")
for idx, manager in enumerate(managers):
    print(f"\nManager {idx+1}:")
    manager.display_info()
```

```python
### Lab Assignment-2: Library Management System
class Book:
    def _init_(self, book_id, title, author):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.is_lent = False

    def display_book(self):
        print(f"ID: {self.book_id}, Title: {self.title}, Author: {self.author}, Status: {'Lent' if self.is_lent else 'Available'}")

class Member:
    def _init_(self, member_id, name):
        self.member_id = member_id
        self.name = name
        self.borrowed_books = []

class Library:
    def _init_(self):
        self.books = []
        self.members = []

    def add_book(self):
        book_id = input("Enter book ID: ")
        title = input("Enter book title: ")
        author = input("Enter author name: ")
        self.books.append(Book(book_id, title, author))
        print("Book added successfully!")

    def lend_book(self):
        member_id = input("Enter member ID: ")
        book_id = input("Enter book ID to lend: ")
        member = next((m for m in self.members if m.member_id == member_id), None)
        book = next((b for b in self.books if b.book_id == book_id and not b.is_lent), None)
        if member and book:
            book.is_lent = True
            member.borrowed_books.append(book)
            print(f"Book '{book.title}' lent to {member.name}.")
        else:
            print("Invalid member or book unavailable.")

    def return_book(self):
        member_id = input("Enter member ID: ")
        book_id = input("Enter book ID to return: ")
        member = next((m for m in self.members if m.member_id == member_id), None)
        book = next((b for b in member.borrowed_books if b.book_id == book_id), None)
        if book:
            book.is_lent = False
            member.borrowed_books.remove(book)
            print(f"Book '{book.title}' returned.")
        else:
            print("Book not found in borrowed list.")

    def display_books(self):
        print("\nLibrary Books:")
        for book in self.books:
            book.display_book()

    def menu(self):
        while True:
            print("\nLibrary Menu:\n1. Add Book\n2. Lend Book\n3. Return Book\n4. Display Books\n5. Exit")
            choice = int(input("Enter choice: "))
            if choice == 1:
                self.add_book()
            elif choice == 2:
                self.lend_book()
            elif choice == 3:
                self.return_book()
            elif choice == 4:
                self.display_books()
            elif choice == 5:
                break
            else:
                print("Invalid choice.")

# Usage
lib = Library()
lib.menu()
```
