```python
#Perform the following operations using Numpy:
#a) Construct a Python program using NumPy to generate a 4 * 4 identity matrix

import numpy as np
identity_matrix = np.eye(4)
print("4x4 Identity Matrix:")
print(identity_matrix)
```

```python
#b) Develop a Python program to generate two 3 * 3 matrices filled with random integers (1 to 9), then perform matrix addition and matrix multiplication.
import numpy as np
A = np.random.randint(1, 10, (3, 3))
B = np.random.randint(1, 10, (3, 3))
print("Matrix A:")
print(A)
print("Matrix B:")
print(B)
# Matrix Addition
addition = A + B
print("Addition of A and B:")
print(addition)
# Matrix Multiplication
multiplication = np.dot(A, B)
print("Multiplication of A and B:")
print(multiplication)
```

```python
#Develop a NumPy program to multiply a 5X3 matrix by a 3X2 matrix and create a product matrix, also print the product matrix
import numpy as np
print("Enter elements for 5x3 matrix:")
matrix1 = []
for i in range(5):
    row = list(map(int, input().split()))
    matrix1.append(row)
matrix1 = np.array(matrix1)
print("Enter elements for 3x2 matrix:")
matrix2 = []
for i in range(3):
    row = list(map(int, input().split()))
    matrix2.append(row)

matrix2 = np.array(matrix2)
# Matrix multiplication
product = np.dot(matrix1, matrix2)

print("Product Matrix (5x2):")
print(product)
```
