```python
# 15. Python program to reverse a given string (5 different ways).
s = "hello"
# 1. slicing
print(s[::-1])
# 2. reversed()
print(''.join(reversed(s)))
# 3. loop
rev = ''.join(s[i] for i in range(len(s)-1, -1, -1))
print(rev)
# 4. recursion
def rev(s):
    return s[-1] + rev(s[:-1]) if s else ''
print(rev(s))
# 5. stack (list)
stack = list(s)
rev = ''.join(stack.pop() for _ in range(len(s)))
print(rev)
```

```python
# 16. Python program to reverse a string using stack and reversed method.
stack = list("hello")
rev = ''.join(stack.pop() for _ in range(len(stack)))
print(rev)
```
