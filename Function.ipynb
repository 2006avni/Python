```python
### Lab Assignment-1
def string_statistics(input_str):
    vowels = 0
    consonants = 0
    spaces = 0
    lowercase = 0

    for char in input_str:
        if char.isalpha():
            if char.lower() in 'aeiou':
                vowels += 1
            else:
                consonants += 1
        if char.isspace():
            spaces += 1
        if char.islower():
            lowercase += 1

    print(f"Number of Vowels: {vowels}")
    print(f"Number of Consonants: {consonants}")
    print(f"Number of Spaces: {spaces}")
    print(f"Number of Lowercase Letters: {lowercase}")

# Main program
input_str = input("Enter a string: ")
string_statistics(input_str)
```

```python
def capitalize_lines(input_lines):
    """Converts all input lines to uppercase."""
    capitalized_lines = [line.upper() for line in input_lines]
    return capitalized_lines

# Main program
def main():
    input_lines = []
    try:
        num_lines = int(input("Enter number of lines: "))
        for i in range(num_lines):
            line = input(f"Enter line {i+1}: ")
            input_lines.append(line)
    except ValueError:
        print("Please enter a valid number.")
        return

    # Capitalize and print lines
    capitalized = capitalize_lines(input_lines)
    print("\nCapitalized Lines:")
    for line in capitalized:
        print(line)

if __name__ == "__main__":
    main()
```
