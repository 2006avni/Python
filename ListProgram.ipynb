```python
#1. Get the largest number from a list.
   num = [12, 45, 7, 23, 56]
   max_num = num[0]
   for n in num:
       if n > max_num:
           max_num = n
   print(max_num)
```

```python
#2. Print numbers of a list after removing even numbers.
   num = [12, 45, 7, 22, 23, 56]
   odd_list = [n for n in num if n % 2 != 0]
   print(odd_list)
```

```python
#3. Reverse a list.
   num = [1, 2, 3, 4, 5]
   rev_list = num[::-1]
   print(rev_list)
```

```python
#4. Square each element of a list.
   num = [1, 2, 3, 4]
   sq_list = [n ** 2 for n in num]
   print(sq_list)
```

```python
#5. Remove an empty element from a list.
   num = ["Hello", 34, 45, "", 40]
   clean_list = [n for n in num if n != ""]
   print(clean_list)
```

```python
#6. Display items from a list that are divisible by 5.
   num = [10, 23, 25, 34, 40]
   div_by_5 = [n for n in num if n % 5 == 0]
   print(div_by_5)
```

```python
#7. Remove repetitive items from a list.
   num = [2, 3, 4, 5, 2, 6, 3, 2]
   unique_list = []
   for n in num:
       if n not in unique_list:
           unique_list.append(n)
   print(unique_list)
```

```python
#8. Append data of the second list to the first list.
   list1 = [23, 24, 25, 26]
   list2 = [27, 28, 29, 30]
   list1 = list1 + list2
   print(list1)
```

```python
#9. Filter odd and even numbers from a list.
   num = [2, 23, 24, 51, 46, 67]
   even = [n for n in num if n % 2 == 0]
   odd = [n for n in num if n % 2 != 0]
   print("Even", even, "Odd", odd)
```

```python
#10. Arrange elements in sorted order.
    num = [34, 12, 45, 7]
    num.sort()
    print(num)
```

```python
#11. Count the number of items stored in a list.
    num = [10, 20, 30, 40]
    count = len(num)
    print(count)
```

```python
#12. Reverse a list of characters using indexing.
    chars = ['a', 'b', 'c', 'd']
    rev_chars = chars[::-1]
    print(rev_chars)
```

```python
#13. Add item 7000 after 6000 in a list.
    list1 = [10, 20, [300, 400, [5000, 6000], 500], 30, 40]
    list1[2][2].insert(1, 7000)
    print(list1)
```

```python
#14. Find value 20 in the list and replace it with 200.
    list1 = [5, 10, 15, 20, 25, 50, 20]
    for i in range(len(list1)):
        if list1[i] == 20:
            list1[i] = 200
            break
    print(list1)
```

```python
#15. Find the range (difference between largest and smallest) of a list.
    num = [34, 12, 45, 7]
    num.sort()
    range_val = num[-1] - num[0]
    print(range_val)
```

```python
#16. Insert a new element at a specific position in the list.
    num = [10, 20, 30, 40]
    num.insert(2, 99)
    print(num)
```

```python
#17. Find common elements between two lists.
    list1 = [1, 2, 3, 4]
    list2 = [3, 4, 5, 6]
    common = [n for n in list1 if n in list2]
    print(common)
```

```python
#18. Find the second smallest element in a list.
    num = [34, 12, 45, 7]
    num.sort()
    second_smallest = num[1]
    print(second_smallest)
```

```python
#19. Remove elements at even indices from a list.
    num = [10, 20, 30, 40, 50]
    new_list = [num[i] for i in range(len(num)) if i % 2 != 0]
    print(new_list)
```

```python
#20. Find the sum of elements at odd indices in a list.
    num = [10, 20, 30, 40, 50]
    odd_sum = sum(num[i] for i in range(len(num)) if i % 2 != 0)
    print(odd_sum)
```
