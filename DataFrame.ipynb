```python
import pandas as pd

# Load dataset
df = pd.read_csv("books.csv")

# a) Print complete report
print("Complete Book Report:")
print(df)

# b) List of books by a given author
author_name = input("Enter author name: ")
author_books = df[df['author'] == author_name]
print("\nBooks by", author_name)
print(author_books)

# c) Books by a given publication
pub_name = input("Enter publication name: ")
pub_books = df[df['publication'] == pub_name]
print("\nBooks by publication:", pub_name)
print(pub_books)

# d) Cheapest and costliest books
cheapest = df[df['price'] == df['price'].min()]
costliest = df[df['price'] == df['price'].max()]

print("\nCheapest Book:")
print(cheapest[['title', 'price']])

print("\nCostliest Book:")
print(costliest[['title', 'price']])

# e) Sort books by year
sorted_books = df.sort_values(by='year')
print("\nBooks sorted by Year:")
print(sorted_books)
```

```python
import pandas as pd

# Create data for 5 states
data = {
    'State': ['Maharashtra', 'Gujarat', 'Rajasthan', 'Tamil Nadu', 'Uttar Pradesh'],
    'Area': [307713, 196244, 342239, 130058, 240928],   # in sq km
    'Population': [124000000, 70000000, 81000000, 78000000, 230000000]
}

df = pd.DataFrame(data)

# a) Print complete information
print("State Information:")
print(df)

# b) State with largest area
largest_area = df[df['Area'] == df['Area'].max()]
print("\nState with Largest Area:")
print(largest_area['State'].values[0])

# c) State with largest population
largest_pop = df[df['Population'] == df['Population'].max()]
print("\nState with Largest Population:")
print(largest_pop['State'].values[0])

# d) Calculate population density
df['Density'] = df['Population'] / df['Area']

print("\nState Data with Population Density:")
print(df)

# e) State with highest population density
highest_density = df[df['Density'] == df['Density'].max()]
print("\nState with Highest Population Density:")
print(highest_density['State'].values[0])
```
